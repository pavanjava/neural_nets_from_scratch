### Coding Multi-Neuron Layer with multiple inputs

```python
X = [1, 2, 3, 2.5]
w = [[0.2, 0.8, -0.5, 1],
     [0.5, -0.91, 0.26, -0.5],
     [-0.26, -0.27, 0.17, 0.87]]
bias = biases = [2, 3, 0.5]

outputs = [
    X[0]*w[0][0] + X[1]*w[0][1] + X[2]*w[0][2] + X[3]*w[0][3] + bias[0],
    X[0]*w[1][0] + X[1]*w[1][1] + X[2]*w[1][2] + X[3]*w[1][3] + bias[1],
    X[0]*w[2][0] + X[1]*w[2][1] + X[2]*w[2][2] + X[3]*w[2][3] + bias[2]
    ]
outputs
```

### using loop to construct the above logic

```python
layer_outputs = []
for weights, _bias in zip(w, bias):
    output = 0
    for _input, weight in zip(X, weights):
        output = output + weight * _input
    output = output + _bias
    layer_outputs.append(output)

layer_outputs
```
