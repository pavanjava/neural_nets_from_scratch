### Coding a layer using numpy

```python
X = [1, 2, 3, 2.5]
w = np.array([[0.2, 0.8, -0.5, 1],
              [0.5, -0.91, 0.26, -0.5],
              [-0.26, -0.27, 0.17, 0.87]])
bias = [2, 3, 0.5]

outputs = np.dot(X, w.T) + bias
outputs
```

### Coding the batch of inputs

```python
inputs = np.array([[1, 2, 3, 2.5],
                   [2, 5, -1, 2.0],
                   [-0.5, 2.2, 3.3, -0.8]])

weights = np.array([[0.2, 0.8, -0.5, 1],
                    [0.5, -0.91, 0.26, -0.5],
                    [-0.26, -0.27, 0.17, 0.87]])

bias = np.array([2.0, 3.0, 0.5])

outputs = np.dot(inputs, weights.T) + bias
outputs
```
